```python
from typing import Annotated, TypedDict
from langgraph.graph.message import add_messages

# GraphState state definition
class GraphState(TypedDict):
    question: Annotated[str, "Question"]  # question
    context: Annotated[str, "Context"]  # result of search
    answer: Annotated[str, "Answer"]  # answer
    messages: Annotated[list, add_messages]  # mesage(accumulated list)
```
